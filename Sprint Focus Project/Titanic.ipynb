```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv("train.csv")
train.head()
```

## Exploração inicial

```python
train.info()
```

```python
train.describe()
```

## Gráficos iniciais

```python
sns.countplot(data=train, x="Survived")
plt.title("Distribuição de sobreviventes")
plt.show()
```

```python
sns.countplot(data=train, x='Sex', hue='Survived')
plt.title("Quantidade de pessoas por sexo no titanic")
plt.show()
```

# Sobrevivência por classe

```python
sns.countplot(data=train, x="Pclass", hue="Survived")
plt.title("Sobrevivência por classe")
plt.show()
```

```python

sns.countplot(data=train[train['Sex']=='female'], x="Pclass", hue="Survived")
plt.title("Sobrevivência por classe")
plt.show()
```

```python

sns.countplot(data=train[train['Sex']=='male'], x="Pclass", hue="Survived")
plt.title("Sobrevivência por classe")
plt.show()
```

```python
# Idade / Sobrevivência

sns.boxplot(data=train, x="Survived", y="Age")
plt.title("Distribuição de idades por sobrevivência")
plt.show()
```

```python
train['Age'].describe()
```

```python
sns.histplot(data=train, x="Age", hue="Survived", bins=20)
plt.title("Distribuição de idades por sobrevivência")
plt.show()
```

```python
sns.boxplot(data=train, x="Pclass", y="Fare", hue="Survived")
plt.show()
```

```python
# Porto de embarque e sobrevivência
sns.countplot(data=train, x="Embarked", hue="Survived")
```
